# Data acquisition

```python
import paho.mqtt.client as mqtt #import the client1


import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm
from datetime import datetime
```

```python
def adquiriDatos():
    broker = "broker.emqx.io"
    port = 1883
    topic = "IMUSensor"
    Qos = 2
    def on_connect(client, userdata, flags, rc):
        if rc == 0:
            print("Connected to the MQTT broker")
            client.subscribe(topic,Qos)
        else:
            print("Error connecting, result code: " + str(rc))

    def on_message(client, userdata, msg):
        current_datetime = datetime.now()
        current_date = current_datetime.strftime('%Y-%m-%d')
        current_time = current_datetime.strftime('%H:%M:%S.%f')
        print(f"Message received on the topic {msg.topic}: {msg.payload.decode()}, {current_date}, {current_time}")
        l_mqtt.append((msg.payload.decode()))
        dates.append(current_date)
        times.append(current_time)
        
    client = mqtt.Client()
    client.on_connect = on_connect
    client.on_message = on_message
    client.connect(broker, port, keepalive=60)
    
    client.loop_start()
    try:
        while True:
            pass
    except KeyboardInterrupt:
        print("Desconectando...")
        client.loop_stop()
        client.disconnect()
```

## Decubitus Supine

```python
l_mqtt = []
dates= []
times = []
adquiriDatos()
```

```python
t_aux = np.zeros((1,9))

for i in l_mqtt:
    t_aux = np.append(t_aux, np.array([eval(i)]), axis=0)
f1 = t_aux.copy()
```

```python
f1 = f1[1:]
```

```python
df1 = pd.DataFrame(f1, columns=('ax','ay','az','gx','gy','gz','mx','my','mz'))
df1['target'] = 0
df1['DateRecord'] = dates
df1['TimeRecord'] = times
df1.head()
```

```python
df1.to_csv('Decubitus_Supine.csv', mode='a', header=False, index=False)
```
